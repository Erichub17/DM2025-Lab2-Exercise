# **Project Report**

##1.1 Preprocessing Steps

For this project, I realized that traditional models and deep learning models need different kinds of input. So, I implemented two specific preprocessing pipelines:

##1. Basic Cleaning (For Naive Bayes & Random Forest)

My goal here was to reduce noise so TF-IDF could work better.

Standardization: I converted all text to lowercase and removed extra whitespaces to make the vocabulary consistent.

Anonymization: I replaced specific user handles (@user) with a generic <USER> token and URLs with <URL>. This prevents the model from overfitting on specific names or links.

Emoji Handling: Instead of removing emojis, I used the emoji library to translate them into text (e.g., 😂 becomes :joy:). This keeps the sentiment information intact.

Hashtags: I appended hashtags to the end of the text to ensure they are treated as important features.

##2. Context-Aware Cleaning (For RoBERTa Competition Model)

For the deep learning model, I wanted to preserve the emotional context (tone) of the tweets.

Emotional Punctuation: I didn't just remove punctuation. I replaced repeated marks like !! with the special token EXCLAIM and ?? with QUESTION. This helps the model understand intensity and confusion.

Entity Handling: I replaced the placeholder [NAME] with PERSON, @mentions with USER, and links with LINK.

Character Normalization: I reduced repeated characters (e.g., sooooo -> soo) to normalize exaggerated spellings often found in social media text.

HTML Decoding: I also used html.unescape to fix broken characters caused by web encoding.

##1.2 Feature Engineering Steps

After cleaning the text, I needed to convert it into numbers so the models could learn from it.

##1. TF-IDF Vectorization

I used TF-IDF (Term Frequency-Inverse Document Frequency) to extract features.

Settings: I set ngram_range=(1, 2), which means I looked at both single words (unigrams) and pairs of words (bigrams). This helps capture phrases like "not happy" which have a different meaning than just "happy".

Feature Limit: I limited the number of features to the top 5,000 to 10,000 most important words to keep the computation efficient while retaining enough information.

##2. High-Dimensional Visualization (t-SNE & UMAP)

To understand the difficulty of this classification task, I used dimensionality reduction techniques to visualize the data.

t-SNE (2D): I projected the TF-IDF features into a 2D map.

UMAP (3D): I also tried a 3D projection using UMAP to see if the clusters would separate better.

Insight: As shown in the graphs I generated, the data points for different emotions are heavily overlapped. There are no clear, separate clusters. This visual proof told me that simple keyword-based features (like TF-IDF) are likely insufficient for distinguishing between complex emotions, which motivated me to use Deep Learning models (RoBERTa) in the later stages.

##1.3 Explanation of Your Model

For the final competition, I decided to move away from traditional machine learning and use Deep Learning Transformers, specifically RoBERTa and DistilBERT, to capture the context of the tweets better.

##1. Model Architecture (Ensemble Approach)

I didn't rely on just one model. I used an Ensemble of two different pre-trained models:

Model A: cardiffnlp/twitter-roberta-base-emotion (Specialized for tweets).

Model B: bhadresh-savani/distilbert-base-uncased-emotion (Faster and lighter).

##2. Training Strategy

5-Fold Cross-Validation: To make sure my model wasn't just "lucky," I split the training data into 5 parts and trained the model 5 times. This gave me a much more stable score.

Class Weights: Since the dataset is imbalanced (too much "Joy"), I added Balanced Class Weights to the loss function. This forces the model to pay more attention to minority classes like "Fear" and "Disgust."

##3. Final Prediction

I combined the predictions from both models using a Weighted Average. The model with the higher validation F1 score got a higher "vote" in the final decision. This pushed my final score to 0.7897, which is way better than the 0.2758 I got from Naive Bayes!

##2. Bonus Section (5 pts Optional)

##2.1 Mention Different Things You Tried

I didn't just stick to the default settings; I tried a few advanced techniques to improve my score:

Custom loss function Focal Loss,i implemented a custom SimpleFocalLoss class in PyTorch . I wanted to see if Focal Loss could handle the hard-to-classify examples better than standard Cross-Entropy. Although I ended up using weighted Cross-Entropy for the final submission, writing the custom loss helped me learn a lot about how PyTorch works.

Dynamic ensemble weights,when combining the models, I didn't just average them (A+B)/2. Instead, I calculated the weights based on their validation performance: Weight = (F1_Score)^2. This allowed the model with better performance in each fold to have more influence on the final result, boosting the ensemble accuracy.

Special token engineering, mentioned in the preprocessing step, I manually added tokens like EXCLAIM and QUESTION to the tokenizer's vocabulary. This was an experiment to see if explicitly telling the model about punctuation intensity would help.

##2.2 Mention Insights You Gained

Naive Bayes (F1: 0.2758): Accuracy is Misleading
My Naive Bayes model achieved a high accuracy of 58.39%, but it was actually failing. The confusion matrix shows it was heavily biased towards the majority class "Joy" (Recall: 93%), while failing completely on "Disgust" (Recall: 0%) and "Sadness" (Recall: 4%). This taught me that accuracy is a terrible metric for imbalanced datasets.

Random Forest (F1: 0.3550): The Trade-off of Fairness
By using class_weight='balanced', the Random Forest model behaved very differently. Its accuracy dropped to 43.29%, but the Macro F1 score improved to 0.3550. Unlike Naive Bayes, it actually tried to learn minority classes (Recall for "Fear" rose to 52%, and "Surprise" to 53%), but at the cost of confusing "Joy" (Recall dropped to 42%) with other emotions.

Deep Learning (F1: 0.7897): Context is King
Neither of the traditional models could truly separate similar emotions like "Anger" and "Joy" effectively because they relied on keywords (TF-IDF). The Deep Learning ensemble was the only approach that understood the context and tone of the sentences, achieving a high F1 score of 0.7897 and solving the classification problem effectively.

```python

# # Step 7 文本清理

import emoji
import re
import pandas as pd

def clean_text_v2(text, hashtags=None):
    """
    基礎文本清理（用於 TF-IDF 和傳統模型）
    """
    if pd.isna(text):
        return ""

    # 1. Emoji 轉文字
    text = emoji.demojize(text, language='en')
    text = text.replace(':', ' ')

    # 2. 保留結構，用特殊 token 替換
    text = re.sub(r'@\w+', '<USER>', text)
    text = re.sub(r'http\S+|www\S+|https\S+', '<URL>', text)

    # 3. 加入 hashtags
    if isinstance(hashtags, list) and len(hashtags) > 0:
        hashtag_text = " ".join(f"#{h}" for h in hashtags if h)
        text = f"{text} {hashtag_text}"

    # 4. 轉小寫
    text = text.lower()

    # 5. 移除多餘空白
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# 清理文本並存入 text_clean
print("\n🔄 清理 Training set...")
train_df['text_clean'] = train_df.apply(
    lambda row: clean_text_v2(row['text'], row.get('hashtags')),
    axis=1
)

print("🔄清理 Test set...")
test_df['text_clean'] = test_df.apply(
    lambda row: clean_text_v2(row['text'], row.get('hashtags')),
    axis=1
)


print(f"  train_df 欄位: {train_df.columns.tolist()}")
print(f"  新增欄位: text_clean")

# 檢查前3筆
print("\n" + "=" * 60)
print("清理前後對比（前 3 筆）")
print("=" * 60)
for i in range(min(3, len(train_df))):
    print(f"\n第 {i+1} 筆")
    print(f"原始文本: {train_df.iloc[i]['text'][:80]}")
    print(f"清理後  : {train_df.iloc[i]['text_clean'][:80]}")


# 對比差異（修正版：只比較原始 vs 新版）

print("=" * 60)
print("清理前後對比（前 3 筆）")
print("=" * 60)
for i in range(min(3, len(train_df))):
    print(f"\n第 {i+1} 筆")
    print(f"原始文本: {train_df.iloc[i]['text'][:100]}")
    print(f"新版清理: {train_df.iloc[i]['text_v2'][:100]}")
    print(f"情緒標籤: {train_df.iloc[i]['emotion']}")

# 檢查文本長度
train_df['text_v2_length'] = train_df['text_v2'].str.split().str.len()
test_df['text_v2_length'] = test_df['text_v2'].str.split().str.len()


print("文本長度統計")

print("\nTraining set:")
print(train_df['text_v2_length'].describe())
print("\nTest set:")
print(test_df['text_v2_length'].describe())

```

```python

# 1. Feature Engineering Steps - TF-IDF

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
import os


print("Step 2.1: TF-IDF 特徵提取")

# 建立資料夾（如果不存在）

os.makedirs('./models', exist_ok=True)
os.makedirs('./Phase3data', exist_ok=True)

print(f"  Training set: {len(train_df)} 筆")
print(f"  Test set: {len(test_df)} 筆")
```

```python
# 2. 建立 TF-IDF Vectorizer


tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2), min_df=2, max_df=0.95)

# 改進建議：
tfidf_vectorizer = TfidfVectorizer(
    max_features=10000,  # 增加特徵數
    ngram_range=(1, 3),  # 加入 trigram
    min_df=2,
    max_df=0.9,  # 降低，過濾更多常見詞
    sublinear_tf=True,  # 使用 log scaling
    strip_accents='unicode',
    lowercase=True
)
```

```python
# 3. 轉換文本為 TF-IDF 特徵

# 用訓練集來 fit，然後 transform 訓練集和測試集
X_train_tfidf = tfidf_vectorizer.fit_transform(train_df['text_clean'])
X_test_tfidf = tfidf_vectorizer.transform(test_df['text_clean'])

# 取得標籤
y_train = train_df['emotion']
y_test = test_df['emotion']

print(f"\n TF-IDF 轉換完成！")
print(f"  訓練集特徵矩陣: {X_train_tfidf.shape}")
print(f"  測試集特徵矩陣: {X_test_tfidf.shape}")
print(f"  特徵數量: {len(tfidf_vectorizer.get_feature_names_out())} 個詞")

```

```python


import os, json, re, html, emoji, gc
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score
from sklearn.utils.class_weight import compute_class_weight
from scipy.special import softmax
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import (
    AutoTokenizer, AutoConfig, AutoModelForSequenceClassification,
    TrainingArguments, Trainer, EarlyStoppingCallback, set_seed,
    DataCollatorWithPadding
)
from datasets import Dataset


# 1. 簡單但不同的文本預處理


def preprocess_text(text, hashtags=None):
    """簡單穩定的預處理"""
    text = html.unescape(str(text))

    # 保留情緒符號（不同策略）
    text = re.sub(r'!{2,}', ' EXCLAIM ', text)
    text = re.sub(r'\?{2,}', ' QUESTION ', text)

    # 標準化
    text = re.sub(r'http\S+|www\.\S+', 'LINK', text)
    text = re.sub(r'@\w+', 'USER', text)
    text = re.sub(r'\[NAME\]', 'PERSON', text, flags=re.IGNORECASE)

    # Emoji 轉換
    text = emoji.demojize(text, language='en')
    text = text.replace(':', ' ')

    # 重複字符（不同方法）
    text = re.sub(r'(.)\1{3,}', r'\1\1', text)

    # Hashtag
    if hashtags and isinstance(hashtags, list):
        tags = ' '.join([str(h).replace('#', '') for h in hashtags if h])
        text = f"{text} {tags}"

    # 清理
    text = re.sub(r'\s+', ' ', text).strip()
    return text



# 2. Focal Loss（

class SimpleFocalLoss(nn.Module):
    def __init__(self, alpha=0.25, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none')
        p_t = torch.exp(-ce_loss)
        loss = self.alpha * (1 - p_t) ** self.gamma * ce_loss
        return loss.mean()


# 3. 自定義 Trainer


# SimpleTrainer
class SimpleTrainer(Trainer):
    def __init__(self, class_weight=None, use_focal=False, **kwargs):
        super().__init__(**kwargs)
        self.class_weight = class_weight
        self.use_focal = use_focal
        if use_focal:
            self.focal_loss = SimpleFocalLoss()

    # 解決 keep_torch_compile 問題
    def _wrap_model(self, model, training=True, dataloader=None):
        if self.args.use_cpu:
            model = model.to(self.args.device)
        if not training:
            return model
        if hasattr(self, 'accelerator'):
            return self.accelerator.prepare(model)
        return model

    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        if self.use_focal:
            loss = self.focal_loss(logits, labels)
        else:
            weight = None
            if self.class_weight is not None:
                weight = torch.tensor(self.class_weight,
                                    dtype=torch.float32,
                                    device=logits.device)
            loss = F.cross_entropy(logits, labels, weight=weight)

        return (loss, outputs) if return_outputs else loss

# 4. 計算 F1 評估指標


def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    f1_macro = f1_score(labels, preds, average='macro')
    f1_micro = f1_score(labels, preds, average='micro')
    return {'f1_macro': f1_macro, 'f1_micro': f1_micro}


# 5. 主訓練函數

def train_single_model(
    model_name,
    train_df,
    test_df,
    label2id,
    n_splits=5,
    lr=2e-5,
    epochs=4,
    batch_size=16,
    max_len=160,
    seed=42
):


    print(f"\n{'='*60}")
    print(f"訓練: {model_name}")
    print(f"{'='*60}\n")

    # Tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)

    # 添加特殊 token
    new_tokens = ['LINK', 'USER', 'PERSON', 'EXCLAIM', 'QUESTION']
    new_tokens = [t for t in new_tokens if t not in tokenizer.get_vocab()]
    if new_tokens:
        tokenizer.add_tokens(new_tokens)
        print(f"新增 tokens: {new_tokens}")

    # 準備標籤
    y = train_df['emotion'].map(label2id).values

    # 初始化
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)
    oof_probs = np.zeros((len(train_df), len(label2id)))
    test_probs = np.zeros((len(test_df), len(label2id)))

    # K-Fold 訓練
    for fold, (tr_idx, va_idx) in enumerate(skf.split(train_df, y), 1):
        print(f"\n--- Fold {fold}/{n_splits} ---")

        # 分割數據
        train_texts = train_df.iloc[tr_idx]['text'].tolist()
        train_labels = y[tr_idx].tolist()
        valid_texts = train_df.iloc[va_idx]['text'].tolist()
        valid_labels = y[va_idx].tolist()
        test_texts = test_df['text'].tolist()

        # Tokenize
        train_encodings = tokenizer(
            train_texts,
            truncation=True,
            padding=False,  # 使用 dynamic padding
            max_length=max_len
        )
        valid_encodings = tokenizer(
            valid_texts,
            truncation=True,
            padding=False,
            max_length=max_len
        )
        test_encodings = tokenizer(
            test_texts,
            truncation=True,
            padding=False,
            max_length=max_len
        )

        # 建立 Dataset
        train_dataset = Dataset.from_dict({
            'input_ids': train_encodings['input_ids'],
            'attention_mask': train_encodings['attention_mask'],
            'label': train_labels
        })
        valid_dataset = Dataset.from_dict({
            'input_ids': valid_encodings['input_ids'],
            'attention_mask': valid_encodings['attention_mask'],
            'label': valid_labels
        })
        test_dataset = Dataset.from_dict({
            'input_ids': test_encodings['input_ids'],
            'attention_mask': test_encodings['attention_mask']
        })

        # 計算類別權重
        class_weights = compute_class_weight(
            'balanced',
            classes=np.unique(train_labels),
            y=train_labels
        )

        # 載入模型
        config = AutoConfig.from_pretrained(
            model_name,
            num_labels=len(label2id),
            hidden_dropout_prob=0.1,
            attention_probs_dropout_prob=0.1
        )

        model = AutoModelForSequenceClassification.from_pretrained(
            model_name,
            config=config,
            ignore_mismatched_sizes=True
        )

        # 調整 embedding 大小
        model.resize_token_embeddings(len(tokenizer))

        # 訓練參數
        training_args = TrainingArguments(
            output_dir=f'./output_fold{fold}',
            learning_rate=lr,
            per_device_train_batch_size=batch_size,
            per_device_eval_batch_size=batch_size * 2,
            num_train_epochs=epochs,
            weight_decay=0.01,
            warmup_ratio=0.1,
            lr_scheduler_type='cosine',
            eval_strategy='epoch',
            save_strategy='epoch',
            load_best_model_at_end=True,
            metric_for_best_model='f1_macro',
            greater_is_better=True,
            save_total_limit=1,
            logging_steps=100,
            bf16=True,
            torch_compile=False,
            report_to='none',
            seed=seed,
            dataloader_num_workers=0,
            dataloader_pin_memory=False
        )

        # Data collator for dynamic padding
        data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

        # Trainer
        trainer = SimpleTrainer(
            model=model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=valid_dataset,
            tokenizer=tokenizer,
            data_collator=data_collator,
            compute_metrics=compute_metrics,
            class_weight=class_weights,
            use_focal=False,  # 可改為 True 使用 Focal Loss
            callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
        )

        # 訓練
        try:
            trainer.train()

            # 預測
            valid_pred = trainer.predict(valid_dataset)
            test_pred = trainer.predict(test_dataset)

            oof_probs[va_idx] = valid_pred.predictions
            test_probs += test_pred.predictions / n_splits

            # 評估
            fold_f1 = f1_score(
                valid_labels,
                np.argmax(valid_pred.predictions, axis=1),
                average='macro'
            )
            print(f"Fold {fold} F1: {fold_f1:.4f}")

        except KeyboardInterrupt:
            print(f"\nFold {fold} 被中斷，跳過...")
            continue
        finally:
            # 清理記憶體
            del model, trainer
            torch.cuda.empty_cache()
            gc.collect()

    # 最終評估
    oof_pred = np.argmax(oof_probs, axis=1)
    final_f1 = f1_score(y, oof_pred, average='macro')
    print(f"\n 整體 OOF F1: {final_f1:.4f}")

    return softmax(oof_probs, axis=1), softmax(test_probs, axis=1)


# 6. 權重平均 Ensemble

def weighted_average_ensemble(all_oof, all_test, y_true):


    weights = []
    for i, oof in enumerate(all_oof):
        pred = np.argmax(oof, axis=1)
        f1 = f1_score(y_true, pred, average='macro')
        weights.append(f1 ** 2)  # 使用平方強化好模型
        print(f"Model {i+1}: F1={f1:.4f}, Weight={f1**2:.4f}")

    # 正規化
    weights = np.array(weights)
    weights = weights / weights.sum()

    # 加權平均
    oof_ensemble = np.average(all_oof, axis=0, weights=weights)
    test_ensemble = np.average(all_test, axis=0, weights=weights)

    final_f1 = f1_score(y_true, np.argmax(oof_ensemble, axis=1), average='macro')
    print(f"\nEnsemble F1: {final_f1:.4f}")

    return oof_ensemble, test_ensemble


# 7. 主程式

if __name__ == "__main__":
    SEED = 42
    set_seed(SEED)


    # train_df, test_df, label2id, id2label

    print("=" * 60)
    print("開始訓練")
    print("=" * 60)

    # 1. 預處理
    print("\n預處理文本...")
    train_df['text'] = train_df.apply(
        lambda r: preprocess_text(r['text'], r.get('hashtags')),
        axis=1
    )
    test_df['text'] = test_df.apply(
        lambda r: preprocess_text(r['text'], r.get('hashtags')),
        axis=1
    )

    # 2. 準備標籤映射
    y = train_df['emotion'].map(label2id).values

    # 3. 訓練多個模型
    models = [
        "cardiffnlp/twitter-roberta-base-emotion",
        "bhadresh-savani/distilbert-base-uncased-emotion"
    ]

    configs = {
        "cardiffnlp/twitter-roberta-base-emotion": {
            "lr": 2e-5, "epochs": 4, "max_len": 160
        },
        "bhadresh-savani/distilbert-base-uncased-emotion": {
            "lr": 3e-5, "epochs": 4, "max_len": 128
        }
    }

    all_oof, all_test = [], []

    for model_name in models:
        cfg = configs[model_name]

        oof, test = train_single_model(
            model_name=model_name,
            train_df=train_df,
            test_df=test_df,
            label2id=label2id,
            n_splits=5,
            lr=cfg['lr'],
            epochs=cfg['epochs'],
            batch_size=8,
            max_len=cfg['max_len'],
            seed=SEED
        )

        all_oof.append(oof)
        all_test.append(test)

    # 4. Ensemble
    print("\n" + "=" * 60)
    print("Ensemble 模型")
    print("=" * 60)

    _, final_test = weighted_average_ensemble(all_oof, all_test, y)

    # 5. csv
    submission = pd.DataFrame({
        'id': test_df['id'],
        'emotion': [id2label[i] for i in np.argmax(final_test, axis=1)]
    })

    submission.to_csv('submission.csv', index=False)
    print("\n 提交檔案已儲存: submission.csv")
```

For the final competition, I decided to move away from traditional machine learning and use Deep Learning Transformers, specifically RoBERTa and DistilBERT.

What I did:

Smarter Preprocessing: Instead of just removing everything, I replaced repeated punctuation like !! with a special token EXCLAIM and user handles with USER. This helps the model understand context (like excitement) rather than just treating them as noise.

5-Fold Cross-Validation: To make sure my model wasn't just 'lucky', I used Stratified K-Fold to train it 5 times on different data splits.

Ensemble Strategy: I trained two different models and combined their predictions using a weighted average. Basically, the model with the higher F1 score got more say in the final decision.

The Results: The improvement is huge!

My single RoBERTa model already hit an F1 score of 0.77.

After combining it with DistilBERT (Ensemble), my final score jumped to 0.7897!

Compared to the 0.28 from Naive Bayes and 0.36 from Random Forest, this proves that for complex emotion detection, we really need models that can 'read' the full sentence structure, not just count keywords.
